```python
import os
import shutil
import pandas as pd

# Path to the extracted ADNI directory
root_dir = "ADNI1_Complete_Extracted/ADNI"
# Path to the CSV file
csv_file = "ADNI1_Filtered_Cleaned_With_Reports.csv"
# Output directory for flattened images
output_dir = "flattened_images"
os.makedirs(output_dir, exist_ok=True)

# Load the CSV file
df = pd.read_csv(csv_file)

# Extract unique image IDs from the CSV
image_ids = df["Image Data ID"].astype(str).unique()

# Function to find and copy NIfTI files
def flatten_images(root_dir, output_dir, image_ids):
    copied_images = 0
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            if file.endswith(".nii"):
                # Extract the image ID from the filename (e.g., I40657.nii)
                image_id = file.split("_")[-1].replace(".nii", "")
                if image_id in image_ids:
                    # Copy the file to the flattened folder
                    src_path = os.path.join(dirpath, file)
                    dst_path = os.path.join(output_dir, f"{image_id}.nii")
                    shutil.copy2(src_path, dst_path)
                    copied_images += 1
    return copied_images

# Flatten the directory
copied_images = flatten_images(root_dir, output_dir, image_ids)
print(f"Total images copied: {copied_images}")

# Check for missing images
missing_images = set(image_ids) - set([f.split(".")[0] for f in os.listdir(output_dir)])
if missing_images:
    print(f"Missing image IDs: {missing_images}")
else:
    print("All images successfully copied.")
```

```python
import os
import nibabel as nib
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Input and output directories
input_dir = "flattened_images"
output_dir = "preprocessed_images"
os.makedirs(output_dir, exist_ok=True)

def preprocess_image(nifti_file):
    """
    Preprocess a single NIfTI file:
    - Normalize intensities
    - Resize to 224x224
    - Enhance contrast using CLAHE
    """
    # Load NIfTI file
    img_data = nib.load(nifti_file).get_fdata()
    
    # Normalize intensities to [0, 1]
    normalized_img = (img_data - img_data.min()) / (img_data.max() - img_data.min())
    
    # Extract the middle axial slice
    middle_slice = normalized_img[:, :, normalized_img.shape[2] // 2]
    
    # Resize to 224x224
    resized_img = cv2.resize(middle_slice, (224, 224), interpolation=cv2.INTER_LINEAR)
    
    # Enhance contrast using CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_img = clahe.apply((resized_img * 255).astype('uint8'))
    
    return resized_img, enhanced_img

# Process all NIfTI files in the input folder
for file in os.listdir(input_dir):
    if file.endswith(".nii"):  # Ensure only NIfTI files are processed
        nifti_path = os.path.join(input_dir, file)
        
        # Preprocess the image
        resized_img, enhanced_img = preprocess_image(nifti_path)
        
        # Save the preprocessed image
        output_path = os.path.join(output_dir, file.replace(".nii", ".png"))
        Image.fromarray(enhanced_img).save(output_path)

# Display a before and after comparison
sample_file = os.path.join(input_dir, os.listdir(input_dir)[0])
resized_img, enhanced_img = preprocess_image(sample_file)

# Plot the original middle slice and preprocessed image
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(resized_img, cmap="gray")
axs[0].set_title("Original Middle Slice (After Normalization & Resizing)")
axs[0].axis("off")

axs[1].imshow(enhanced_img, cmap="gray")
axs[1].set_title("Preprocessed Image (After CLAHE)")
axs[1].axis("off")

plt.tight_layout()
plt.show()

print(f"Preprocessed images saved to {output_dir}")

```

```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from transformers import AutoImageProcessor, AutoModelForImageClassification

# ✅ Check CUDA availability
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths to datasets and preprocessed images
train_csv = "train_data.csv"
val_csv = "val_data.csv"
test_csv = "test_data.csv"
image_dir = "preprocessed_images"

# Load Data Splits
train_data = pd.read_csv(train_csv)
val_data = pd.read_csv(val_csv)
test_data = pd.read_csv(test_csv)

# ✅ Load ViT Model and Image Processor (Now in PyTorch)
checkpoint = "google/vit-base-patch16-224-in21k"
image_processor = AutoImageProcessor.from_pretrained(checkpoint)
model = AutoModelForImageClassification.from_pretrained(
    checkpoint, num_labels=3  # Number of classes: CN, MCI, AD
).to(device)  # ✅ Move model to CUDA

# ✅ Define PyTorch Dataset
class MRIImageDataset(Dataset):
    def __init__(self, dataframe, image_dir, transform=None):
        self.dataframe = dataframe
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, index):
        row = self.dataframe.iloc[index]
        img_path = os.path.join(self.image_dir, f"{row['Image Data ID']}.png")

        # Load Image
        image = Image.open(img_path).convert("RGB")

        # Apply transforms
        if self.transform:
            image = self.transform(image)

        label = int(row["Group"])  # Convert label to integer
        return image, label

# ✅ Define Transformations for PyTorch
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Ensure image size matches ViT input
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])  # Normalize as required by ViT
])

# ✅ Create PyTorch Dataloaders (with CUDA support)
batch_size = 8  # Adjust based on memory
train_dataset = MRIImageDataset(train_data, image_dir, transform=transform)
val_dataset = MRIImageDataset(val_data, image_dir, transform=transform)
test_dataset = MRIImageDataset(test_data, image_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

# ✅ Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=2e-5)

# ✅ Training Loop (With CUDA)
epochs = 5
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)  # ✅ Move data to CUDA

        optimizer.zero_grad()
        outputs = model(images).logits  # Forward pass
        loss = criterion(outputs, labels)  # Compute loss
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
        total += labels.size(0)

    train_acc = correct / total
    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}, Train Acc: {train_acc:.4f}")

    # ✅ Validation Step
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images).logits
            _, predicted = outputs.max(1)
            correct += predicted.eq(labels).sum().item()
            total += labels.size(0)

    val_acc = correct / total
    print(f"Validation Accuracy: {val_acc:.4f}")

# ✅ Evaluate Model on Test Set
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images).logits
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
        total += labels.size(0)

test_acc = correct / total
print(f"🔥 Test Accuracy: {test_acc:.4f}")
```

```python
import os
import pandas as pd
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
from sklearn.preprocessing import LabelEncoder

# ✅ Set device (Use CUDA if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ✅ Load Data
train_df = pd.read_csv("train_data.csv")
val_df = pd.read_csv("val_data.csv")
test_df = pd.read_csv("test_data.csv")

# Extract text and labels
train_texts = train_df["Text_Report"].tolist()
val_texts = val_df["Text_Report"].tolist()
test_texts = test_df["Text_Report"].tolist()

train_labels = train_df["Group"].tolist()
val_labels = val_df["Group"].tolist()
test_labels = test_df["Group"].tolist()

# ✅ Encode Labels
label_encoder = LabelEncoder()
label_encoder.fit(train_labels + val_labels + test_labels)

train_label_ids = label_encoder.transform(train_labels)
val_label_ids = label_encoder.transform(val_labels)
test_label_ids = label_encoder.transform(test_labels)

# ✅ Load RoBERTa Model (PyTorch Version)
model_name = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
roberta_model = AutoModel.from_pretrained(model_name).to(device)
roberta_model.eval()  # Set model to evaluation mode

# ✅ Function to Generate RoBERTa Embeddings
def get_roberta_embeddings(text_list, tokenizer, model, batch_size=8, max_length=512):
    """
    Encodes a list of texts with RoBERTa (PyTorch version),
    returning an [N, hidden_dim] array via mean pooling of the last hidden states.
    """
    all_embeddings = []
    
    # Use DataLoader for efficient batching
    text_loader = DataLoader(text_list, batch_size=batch_size, shuffle=False)
    
    with torch.no_grad():
        for batch_texts in text_loader:
            encodings = tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                max_length=max_length,
                return_tensors="pt"
            ).to(device)  # Move to GPU

            # Get model outputs
            outputs = model(**encodings)
            last_hidden_state = outputs.last_hidden_state  # Shape: [batch_size, seq_len, hidden_dim]

            # Compute mean pooling across seq_len
            batch_emb = last_hidden_state.mean(dim=1)  # Shape: [batch_size, hidden_dim]

            # Convert to numpy and append
            all_embeddings.append(batch_emb.cpu().numpy())

    return np.concatenate(all_embeddings, axis=0)

# ✅ Generate RoBERTa Embeddings for Text Reports
train_embeddings = get_roberta_embeddings(train_texts, tokenizer, roberta_model, batch_size=8, max_length=512)
val_embeddings = get_roberta_embeddings(val_texts, tokenizer, roberta_model, batch_size=8, max_length=512)
test_embeddings = get_roberta_embeddings(test_texts, tokenizer, roberta_model, batch_size=8, max_length=512)

# ✅ Save embeddings and labels as `.npy` files
np.save("train_text_embeddings.npy", train_embeddings)
np.save("val_text_embeddings.npy", val_embeddings)
np.save("test_text_embeddings.npy", test_embeddings)

np.save("train_text_labels.npy", train_label_ids)
np.save("val_text_labels.npy", val_label_ids)
np.save("test_text_labels.npy", test_label_ids)

print("✅ RoBERTa embeddings extracted and saved successfully!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# ✅ Load Extracted ViT and RoBERTa Features
train_vit = np.load("train_vit_features.npy")
val_vit = np.load("val_vit_features.npy")
test_vit = np.load("test_vit_features.npy")

train_roberta = np.load("train_text_embeddings.npy")
val_roberta = np.load("val_text_embeddings.npy")
test_roberta = np.load("test_text_embeddings.npy")

# ✅ Load Labels
train_labels = np.load("train_text_labels.npy")
val_labels = np.load("val_text_labels.npy")
test_labels = np.load("test_text_labels.npy")

# ✅ Convert Data to PyTorch Tensors
train_vit_tensor = torch.tensor(train_vit, dtype=torch.float32)
val_vit_tensor = torch.tensor(val_vit, dtype=torch.float32)
test_vit_tensor = torch.tensor(test_vit, dtype=torch.float32)

train_roberta_tensor = torch.tensor(train_roberta, dtype=torch.float32)
val_roberta_tensor = torch.tensor(val_roberta, dtype=torch.float32)
test_roberta_tensor = torch.tensor(test_roberta, dtype=torch.float32)

train_labels_tensor = torch.tensor(train_labels, dtype=torch.long)
val_labels_tensor = torch.tensor(val_labels, dtype=torch.long)
test_labels_tensor = torch.tensor(test_labels, dtype=torch.long)

# ✅ Create DataLoaders
batch_size = 32
train_dataset = TensorDataset(train_vit_tensor, train_roberta_tensor, train_labels_tensor)
val_dataset = TensorDataset(val_vit_tensor, val_roberta_tensor, val_labels_tensor)
test_dataset = TensorDataset(test_vit_tensor, test_roberta_tensor, test_labels_tensor)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# ✅ Define Mid-Fusion MLP Model
class MidFusionMLP(nn.Module):
    def __init__(self, vit_input_dim, roberta_input_dim, num_classes):
        super(MidFusionMLP, self).__init__()

        # Reduce ViT features from 768 → 128
        self.vit_fc = nn.Sequential(
            nn.Linear(vit_input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        # Reduce RoBERTa features from 768 → 384
        self.roberta_fc = nn.Sequential(
            nn.Linear(roberta_input_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        # Concatenate and pass through final classifier
        self.final_fc = nn.Sequential(
            nn.Linear(128 + 512, 384),  # 128 (ViT) + 384 (RoBERTa) = 512
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(384, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, vit_x, roberta_x):
        vit_x = self.vit_fc(vit_x)
        roberta_x = self.roberta_fc(roberta_x)
        fused_x = torch.cat((vit_x, roberta_x), dim=1)
        return self.final_fc(fused_x)

# ✅ Set Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ✅ Initialize Mid-Fusion MLP Model
num_classes = len(np.unique(train_labels))
model = MidFusionMLP(vit_input_dim=768, roberta_input_dim=768, num_classes=num_classes).to(device)

# ✅ Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# ✅ Training Loop with Early Stopping
epochs = 50
patience = 5
best_val_loss = float("inf")
counter = 0

for epoch in range(epochs):
    model.train()
    train_loss = 0.0
    correct = 0
    total = 0

    for vit_inputs, roberta_inputs, labels in train_loader:
        vit_inputs, roberta_inputs, labels = vit_inputs.to(device), roberta_inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(vit_inputs, roberta_inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    train_acc = correct / total
    avg_train_loss = train_loss / len(train_loader)

    # Validation Step
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for vit_inputs, roberta_inputs, labels in val_loader:
            vit_inputs, roberta_inputs, labels = vit_inputs.to(device), roberta_inputs.to(device), labels.to(device)
            outputs = model(vit_inputs, roberta_inputs)
            loss = criterion(outputs, labels)

            val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    val_acc = correct / total
    avg_val_loss = val_loss / len(val_loader)

    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.4f}, "
          f"Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.4f}")

    # Early Stopping
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        counter = 0
        torch.save(model.state_dict(), "best_mlp_midfusion.pth")  # Save best model
        print(f"✅ Model saved at epoch {epoch+1} with lowest validation loss: {avg_val_loss:.4f}")
    else:
        counter += 1
        if counter >= patience:
            print("❌ Early stopping triggered!")
            break

# ✅ Load the Best Mid-Fusion MLP Model for Testing
model.load_state_dict(torch.load("best_mlp_midfusion.pth"))
model.to(device)
print(f"✅ Best model loaded from best_mlp_midfusion.pth")

# ✅ Evaluate MLP on Test Set
model.eval()
correct = 0
total = 0
all_preds = []
all_labels = []

with torch.no_grad():
    for vit_inputs, roberta_inputs, labels in test_loader:
        vit_inputs, roberta_inputs, labels = vit_inputs.to(device), roberta_inputs.to(device), labels.to(device)
        outputs = model(vit_inputs, roberta_inputs)
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

test_acc = correct / total
print(f"🔥 Final Test Accuracy (MLP with Mid-Fusion): {test_acc:.4f}")
print("Classification Report (MLP):")
print(classification_report(all_labels, all_preds))

```

```python
import os
import shutil
import pandas as pd

# Path to the extracted ADNI directory
root_dir = "ADNI1_Complete_Extracted"
# Path to the CSV file
csv_file = "ADNI1_Final_With_Biomarkers.csv"
# Output directory for flattened images
output_dir = "flattened_images_2"
os.makedirs(output_dir, exist_ok=True)

# Load the CSV file
df = pd.read_csv(csv_file)

# Extract unique image IDs from the CSV
image_ids = df["Image Data ID"].astype(str).unique()

# Function to find and copy NIfTI files
def flatten_images(root_dir, output_dir, image_ids):
    copied_images = 0
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            if file.endswith(".nii"):
                # Extract the image ID from the filename (e.g., I40657.nii)
                image_id = file.split("_")[-1].replace(".nii", "")
                if image_id in image_ids:
                    # Copy the file to the flattened folder
                    src_path = os.path.join(dirpath, file)
                    dst_path = os.path.join(output_dir, f"{image_id}.nii")
                    shutil.copy2(src_path, dst_path)
                    copied_images += 1
    return copied_images

# Flatten the directory
copied_images = flatten_images(root_dir, output_dir, image_ids)
print(f"Total images copied: {copied_images}")

# Check for missing images
missing_images = set(image_ids) - set([f.split(".")[0] for f in os.listdir(output_dir)])
if missing_images:
    print(f"Missing image IDs: {missing_images}")
else:
    print("All images successfully copied.")
```

```python
import os
import nibabel as nib
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Input and output directories
input_dir = "flattened_images_2"
output_dir = "preprocessed_images_2"
os.makedirs(output_dir, exist_ok=True)

def preprocess_image(nifti_file):
    """
    Preprocess a single NIfTI file:
    - Normalize intensities
    - Resize to 224x224
    - Enhance contrast using CLAHE
    """
    # Load NIfTI file
    img_data = nib.load(nifti_file).get_fdata()
    
    # Normalize intensities to [0, 1]
    normalized_img = (img_data - img_data.min()) / (img_data.max() - img_data.min())
    
    # Extract the middle axial slice
    middle_slice = normalized_img[:, :, normalized_img.shape[2] // 2]
    
    # Resize to 224x224
    resized_img = cv2.resize(middle_slice, (224, 224), interpolation=cv2.INTER_LINEAR)
    
    # Enhance contrast using CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_img = clahe.apply((resized_img * 255).astype('uint8'))
    
    return resized_img, enhanced_img

# Process all NIfTI files in the input folder
for file in os.listdir(input_dir):
    if file.endswith(".nii"):  # Ensure only NIfTI files are processed
        nifti_path = os.path.join(input_dir, file)
        
        # Preprocess the image
        resized_img, enhanced_img = preprocess_image(nifti_path)
        
        # Save the preprocessed image
        output_path = os.path.join(output_dir, file.replace(".nii", ".png"))
        Image.fromarray(enhanced_img).save(output_path)

# Display a before and after comparison
sample_file = os.path.join(input_dir, os.listdir(input_dir)[0])
resized_img, enhanced_img = preprocess_image(sample_file)

# Plot the original middle slice and preprocessed image
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(resized_img, cmap="gray")
axs[0].set_title("Original Middle Slice (After Normalization & Resizing)")
axs[0].axis("off")

axs[1].imshow(enhanced_img, cmap="gray")
axs[1].set_title("Preprocessed Image (After CLAHE)")
axs[1].axis("off")

plt.tight_layout()
plt.show()

print(f"Preprocessed images saved to {output_dir}")

```

```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from transformers import DeiTFeatureExtractor, DeiTForImageClassification
from torch.optim.lr_scheduler import ReduceLROnPlateau

scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=2, verbose=True)


# ✅ Check CUDA availability
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths to datasets and preprocessed images
train_csv = "train_data_2.csv"
val_csv = "val_data_2.csv"
test_csv = "test_data_2.csv"
image_dir = "preprocessed_images_2"

# Load Data Splits
train_data = pd.read_csv(train_csv)
val_data = pd.read_csv(val_csv)
test_data = pd.read_csv(test_csv)

# ✅ Load DeiT Model and Feature Extractor
checkpoint = "facebook/deit-base-distilled-patch16-224"
feature_extractor = DeiTFeatureExtractor.from_pretrained(checkpoint)
model = DeiTForImageClassification.from_pretrained(
    checkpoint, num_labels=3  # Number of classes: CN, MCI, AD
).to(device)  # ✅ Move model to CUDA


# Define label mapping
label_map = {"CN": 0, "MCI": 1, "AD": 2}

class MRIImageDataset(Dataset):
    def __init__(self, dataframe, image_dir, transform=None):
        self.dataframe = dataframe
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, index):
        row = self.dataframe.iloc[index]
        img_path = os.path.join(self.image_dir, f"{row['Image Data ID']}.png")

        # Load Image
        image = Image.open(img_path).convert("RGB")

        # Apply transforms
        if self.transform:
            image = self.transform(image)

        # Convert string label ('CN', 'MCI', 'AD') to integer (0, 1, 2)
        label = label_map[row["Group"]]  
        return image, label


# ✅ Define Transformations for DeiT
transform = transforms.Compose([
    transforms.ToTensor(),  # Converts NumPy image to PyTorch Tensor
    transforms.Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)  # Ensure correct input scale for DeiT
])


# ✅ Create PyTorch Dataloaders
batch_size = 8  # Adjust based on memory
train_dataset = MRIImageDataset(train_data, image_dir, transform=transform)
val_dataset = MRIImageDataset(val_data, image_dir, transform=transform)
test_dataset = MRIImageDataset(test_data, image_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

# ✅ Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=2e-5)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=2, verbose=True)

# ✅ Training Loop with Early Stopping
epochs = 10
best_val_loss = float("inf")
best_model_state = None
patience = 3
wait = 0

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images).logits  
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
        total += labels.size(0)

    train_loss = running_loss / len(train_loader)
    train_acc = correct / total

    # ✅ Validation Step
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images).logits
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            correct += predicted.eq(labels).sum().item()
            total += labels.size(0)

    val_loss /= len(val_loader)
    val_acc = correct / total

    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
    scheduler.step(val_loss)

    # ✅ Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_model_state = model.state_dict()
        wait = 0
    else:
        wait += 1
        if wait >= patience:
            print("Early stopping triggered.")
            break

# ✅ Load Best Model
if best_model_state:
    model.load_state_dict(best_model_state)

# ✅ Evaluate Model on Test Set
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images).logits
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
        total += labels.size(0)

test_acc = correct / total
print(f"🔥 Test Accuracy: {test_acc:.4f}")


```

```python
import numpy as np
import torch

def extract_embeddings(model, X_np):
    """
    Extract intermediate embeddings from a trained TabNetClassifier.
    
    This function registers a forward hook on the final mapping of the underlying 
    TabNet (stored in model.network.tabnet.final_mapping) to capture its input, which
    is considered the learned embedding, during a forward pass.
    
    Parameters
    ----------
    model : TabNetClassifier
        A trained TabNet model.
    X_np : np.array
        Input data as a NumPy array.
    
    Returns
    -------
    embeddings_array : np.array
        Extracted embeddings for all samples.
    """
    # Set the underlying network to evaluation mode.
    model.network.eval()
    
    embeddings_list = []
    
    # Define a hook function that captures the input to the final mapping.
    def hook_fn(module, input, output):
        # input[0] is the tensor entering the final mapping layer.
        embeddings_list.append(input[0].detach().cpu().numpy())
    
    # Register the hook on the final mapping layer.
    hook_handle = model.network.tabnet.final_mapping.register_forward_hook(hook_fn)
    
    # Run a forward pass over X_np.
    _ = model.predict(X_np)
    
    # Remove the hook.
    hook_handle.remove()
    
    # Concatenate the embeddings from all batches.
    embeddings_array = np.concatenate(embeddings_list, axis=0)
    return embeddings_array

# === Usage Example ===
# (Assuming your model has already been trained and X_train_np, X_val_np, X_test_np exist)

train_embeddings = extract_embeddings(model, X_train_np)
val_embeddings   = extract_embeddings(model, X_val_np)
test_embeddings  = extract_embeddings(model, X_test_np)

print("Train embeddings shape:", train_embeddings.shape)
print("Validation embeddings shape:", val_embeddings.shape)
print("Test embeddings shape:", test_embeddings.shape)
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, TensorDataset
from sklearn.metrics import accuracy_score, classification_report
import os

# ----------------------------------------------------------------------
# 1. LOAD DATA
#    Adjust paths/names as needed. We'll assume you have:
#    - DeiT features (768-dim) as .npy
#    - TabNet features (arbitrary dim) as e.g. train_embeddings_2, val_embeddings_2, test_embeddings_2
#    - Labels y_train, y_val, y_test
# ----------------------------------------------------------------------
# Example: Load DeiT features from .npy
train_deit = np.load("Deit_train_features.npy")  # shape: (N_train, 768)
val_deit   = np.load("Deit_val_features.npy")    # shape: (N_val, 768)
test_deit  = np.load("Deit_test_features.npy")   # shape: (N_test, 768)

# Suppose you already have your TabNet (or other tabular) features in memory,
# e.g. train_embeddings_2, val_embeddings_2, test_embeddings_2.
# They might be NumPy arrays of shape (N, tab_dim).
# Adjust these variable names to match your code.
train_tab = train_embeddings_2  # shape: (N_train, tab_dim)
val_tab   = val_embeddings_2    # shape: (N_val,   tab_dim)
test_tab  = test_embeddings_2   # shape: (N_test,  tab_dim)

# Suppose you also have your labels in NumPy arrays: y_train, y_val, y_test
# shape: (N_train,), (N_val,), (N_test,), integer-encoded for classification
# e.g., 0..num_classes-1
print("train_deit shape:", train_deit.shape)
print("train_tab shape: ", train_tab.shape)
print("y_train shape:   ", y_train.shape)

# ----------------------------------------------------------------------
# 2. CREATE DATASETS & DATALOADERS
#    We'll wrap the DeiT features, TabNet features, and labels in a Dataset.
# ----------------------------------------------------------------------
# Convert data to Torch tensors
train_deit_t = torch.tensor(train_deit, dtype=torch.float32)
val_deit_t   = torch.tensor(val_deit,   dtype=torch.float32)
test_deit_t  = torch.tensor(test_deit,  dtype=torch.float32)

train_tab_t  = torch.tensor(train_tab,  dtype=torch.float32)
val_tab_t    = torch.tensor(val_tab,    dtype=torch.float32)
test_tab_t   = torch.tensor(test_tab,   dtype=torch.float32)

y_train_t    = torch.tensor(y_train,    dtype=torch.long)
y_val_t      = torch.tensor(y_val,      dtype=torch.long)
y_test_t     = torch.tensor(y_test,     dtype=torch.long)

# Build TensorDatasets that return (deit_feats, tab_feats, label)
train_dataset = TensorDataset(train_deit_t, train_tab_t, y_train_t)
val_dataset   = TensorDataset(val_deit_t,   val_tab_t,   y_val_t)
test_dataset  = TensorDataset(test_deit_t,  test_tab_t,  y_test_t)

batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

# ----------------------------------------------------------------------
# 3. DEFINE A MID-FUSION MODEL
#    We'll have:
#    - An image branch (768 -> 384)
#    - A tab branch (identity or small transform)
#    - Concat
#    - Final MLP for classification
# ----------------------------------------------------------------------
class MidFusionModel(nn.Module):
    def __init__(self, 
                 image_dim=768,    # original DeiT dimension
                 tab_dim=8,        # original TabNet (tabular) dimension
                 tab_proj_dim=64,  # expanded dimension for tabular features
                 reduced_dim=384,  # reduced dimension for image features
                 hidden_dim=64, 
                 num_classes=3):
        super(MidFusionModel, self).__init__()
        
        # Branch 1: Image -> project 768 -> 384
        self.image_branch = nn.Sequential(
            nn.Linear(image_dim, reduced_dim),
            nn.ReLU()
        )
        
        # Branch 2: Tab -> project 8 -> 32
        self.tab_branch = nn.Sequential(
            nn.Linear(tab_dim, tab_proj_dim),
            nn.ReLU()
        )
        
        # After fusion, we have (384 + 32) dimensions
        fusion_dim = reduced_dim + tab_proj_dim
        
        # Final classifier
        self.classifier = nn.Sequential(
            nn.Linear(fusion_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, num_classes)
        )
    
    def forward(self, img_feats, tab_feats):
        # 1. Transform each modality
        x_img = self.image_branch(img_feats)     # shape: (batch, 384)
        x_tab = self.tab_branch(tab_feats)       # shape: (batch, 32)

        # 2. Concatenate
        x_fused = torch.cat((x_img, x_tab), dim=1)  # (batch, 416)

        # 3. Classify
        logits = self.classifier(x_fused)
        return logits

# Create the model
image_dim   = train_deit.shape[1]          # 768
tab_dim     = train_tab.shape[1]           # e.g. 32 or whatever your tabular dimension is
reduced_dim = 128                          # dimension to which we reduce image features
hidden_dim  = 64                           # hidden layer in final MLP
num_classes = len(np.unique(y_train))      # or manually set if known

model = MidFusionModel(
    image_dim=image_dim,
    tab_dim=tab_dim,
    reduced_dim=reduced_dim,
    hidden_dim=hidden_dim,
    num_classes=num_classes
)

# ----------------------------------------------------------------------
# 4. TRAINING SETUP
# ----------------------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# Early Stopping parameters
num_epochs = 50
best_val_loss = float('inf')
patience = 10
patience_counter = 0

# ----------------------------------------------------------------------
# 5. TRAINING LOOP with EARLY STOPPING
# ----------------------------------------------------------------------
for epoch in range(num_epochs):
    # ---- TRAIN ----
    model.train()
    running_loss = 0.0
    for deit_batch, tab_batch, labels in train_loader:
        deit_batch, tab_batch, labels = (
            deit_batch.to(device),
            tab_batch.to(device),
            labels.to(device)
        )
        
        optimizer.zero_grad()
        outputs = model(deit_batch, tab_batch)  # forward pass
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * deit_batch.size(0)
    
    train_loss = running_loss / len(train_loader.dataset)
    
    # ---- VALIDATION ----
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for deit_batch, tab_batch, labels in val_loader:
            deit_batch, tab_batch, labels = (
                deit_batch.to(device),
                tab_batch.to(device),
                labels.to(device)
            )
            outputs = model(deit_batch, tab_batch)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * deit_batch.size(0)
            
            _, preds = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (preds == labels).sum().item()
    
    val_loss = val_loss / len(val_loader.dataset)
    val_acc = correct / total
    
    print(f"Epoch {epoch+1}/{num_epochs} "
          f"- Train Loss: {train_loss:.4f} "
          f"- Val Loss: {val_loss:.4f} "
          f"- Val Acc: {val_acc:.4f}")
    
    # Early Stopping logic
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), "best_mid_fusion_model.pth")
    else:
        patience_counter += 1
    
    if patience_counter >= patience:
        print("Early stopping triggered.")
        break

# ----------------------------------------------------------------------
# 6. LOAD BEST MODEL & EVALUATE ON TEST SET
# ----------------------------------------------------------------------
model.load_state_dict(torch.load("best_mid_fusion_model.pth", map_location=device))
model.eval()

all_preds = []
all_labels = []
with torch.no_grad():
    for deit_batch, tab_batch, labels in test_loader:
        deit_batch, tab_batch, labels = (
            deit_batch.to(device),
            tab_batch.to(device),
            labels.to(device)
        )
        outputs = model(deit_batch, tab_batch)
        _, preds = torch.max(outputs, 1)
        
        all_preds.append(preds.cpu().numpy())
        all_labels.append(labels.cpu().numpy())

test_preds = np.concatenate(all_preds, axis=0)
test_labels = np.concatenate(all_labels, axis=0)

acc = accuracy_score(test_labels, test_preds)
print("\nTest Accuracy:", acc)

print("Classification Report (Test):")
unique_classes = np.unique(test_labels)  # or use label_encoder.classes_ if you have one
target_names = [str(c) for c in unique_classes]
print(classification_report(test_labels, test_preds, target_names=target_names))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Assume filtered_df is already defined with the columns:
# ['Age', 'GENOTYPE', 'CDGLOBAL', 'CDRSB', 'MMSCORE', 'NPISCORE', 'GDTOTAL', 'ABETA42', 'TAU', 'PTAU', 'Group']
# For example:
# filtered_df = pd.read_csv("your_filtered_file.csv")

# Define features and label
features = ['Age', 'GENOTYPE', 'CDGLOBAL', 'CDRSB', 'MMSCORE', 'NPISCORE', 'GDTOTAL', 'ABETA42', 'TAU', 'PTAU']
label = 'Group'

# Split the dataset into training (70%), validation (15%), and test (15%)
train_df, temp_df = train_test_split(filtered_df, test_size=0.3, random_state=42, stratify=filtered_df[label])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df[label])
print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))

def generate_textual_report(row):
    """
    Generates a descriptive and neutral clinical report from a row of clinical data,
    using thresholds for each feature. If a feature is missing, it is skipped.
    Age is simply appended as a number.
    """
    report_parts = []

    # Age: simply include the numeric value.
    age = row.get('Age')
    if pd.notnull(age):
        report_parts.append(f"The patient is {age} years old")
    
    # GENOTYPE: Describe risk based on genotype.
    genotype = row.get('GENOTYPE')
    if pd.notnull(genotype):
        if genotype == "4/4":
            report_parts.append("and has a high risk genotype")
        elif genotype == "3/3":
            report_parts.append("and has a typical genotype")
        else:
            report_parts.append("and has an intermediate risk genotype")
    
    # CDGLOBAL: Use thresholds based on common clinical interpretations.
    cdglobal = row.get('CDGLOBAL')
    if pd.notnull(cdglobal):
        if cdglobal <= 0:
            report_parts.append("with no evident global cognitive decline")
        elif cdglobal < 1:
            report_parts.append("with minimal global cognitive alterations")
        elif cdglobal < 2:
            report_parts.append("with mild global cognitive changes")
        elif cdglobal < 3:
            report_parts.append("with moderate global cognitive decline")
        else:
            report_parts.append("with severe global cognitive decline")
    
    # CDRSB: Thresholds: <1: minimal, 1-4: mild, 4-7: moderate, >7: severe.
    cdrsb = row.get('CDRSB')
    if pd.notnull(cdrsb):
        if cdrsb < 1:
            report_parts.append("exhibits minimal impairment on the CDR Sum of Boxes")
        elif cdrsb < 4:
            report_parts.append("shows mild impairment on the CDR Sum of Boxes")
        elif cdrsb < 7:
            report_parts.append("demonstrates moderate impairment on the CDR Sum of Boxes")
        else:
            report_parts.append("displays severe impairment on the CDR Sum of Boxes")
    
    # MMSCORE (MMSE): Typically, >=28 is normal, 24-27 is mild, <24 is significant.
    mmscore = row.get('MMSCORE')
    if pd.notnull(mmscore):
        if mmscore >= 28:
            report_parts.append("with normal cognitive performance on the MMSE")
        elif mmscore >= 24:
            report_parts.append("with mild cognitive impairment on the MMSE")
        else:
            report_parts.append("with significant cognitive deficits on the MMSE")
    
    # NPISCORE: We assume a scale where 0 means no impairment.
    npiscore = row.get('NPISCORE')
    if pd.notnull(npiscore):
        if npiscore == 0:
            report_parts.append("and no notable neuropsychological deficits")
        elif npiscore < 5:
            report_parts.append("with minimal neuropsychological deficits")
        elif npiscore < 10:
            report_parts.append("with mild neuropsychological deficits")
        elif npiscore < 15:
            report_parts.append("with moderate neuropsychological deficits")
        else:
            report_parts.append("with severe neuropsychological deficits")
    
    # GDTOTAL: A global dysfunction score; assume lower values (<3) are minimal, 3-6 mild, 6-9 moderate, >9 severe.
    gdtotal = row.get('GDTOTAL')
    if pd.notnull(gdtotal):
        if gdtotal < 3:
            report_parts.append("with minimal global dysfunction")
        elif gdtotal < 6:
            report_parts.append("with mild global dysfunction")
        elif gdtotal < 9:
            report_parts.append("with moderate global dysfunction")
        else:
            report_parts.append("with severe global dysfunction")
    
    # ABETA42: Lower levels are concerning; assume <250 is low, otherwise normal.
    abeta42 = row.get('ABETA42')
    if pd.notnull(abeta42):
        if abeta42 < 250:
            report_parts.append("and exhibits low amyloid-beta levels")
        else:
            report_parts.append("and shows amyloid-beta levels within the normal range")
    
    # TAU: Higher levels (>50) are concerning.
    tau = row.get('TAU')
    if pd.notnull(tau):
        if tau > 50:
            report_parts.append("with elevated tau levels")
        else:
            report_parts.append("with tau levels within normal limits")
    
    # PTAU: Higher levels (>15) are concerning.
    ptau = row.get('PTAU')
    if pd.notnull(ptau):
        if ptau > 15:
            report_parts.append("and elevated phosphorylated tau levels")
        else:
            report_parts.append("and phosphorylated tau levels are normal")
    
    # Combine all parts into a single report string.
    report = ". ".join(report_parts) + "."
    return report

def add_textual_reports(df):
    """
    Iterates over each row in the DataFrame and generates a textual report.
    The report is then appended as a new column 'textual_report'.
    """
    reports = []
    for index, row in df.iterrows():
        report = generate_textual_report(row)
        reports.append(report)
    df["textual_report"] = reports
    return df

# Process each split to create the textual reports.
print("Generating textual reports for training data...")
train_df = add_textual_reports(train_df)
print("Generating textual reports for validation data...")
val_df = add_textual_reports(val_df)
print("Generating textual reports for test data...")
test_df = add_textual_reports(test_df)

# Optionally, save the resulting DataFrames to CSV files.
train_df.to_csv("train_with_textual_reports_1.csv", index=False)
val_df.to_csv("val_with_textual_reports_1.csv", index=False)
test_df.to_csv("test_with_textual_reports_1.csv", index=False)

print("Textual report generation complete. Reports have been appended to each split.")
```

```python
import os
import pandas as pd
import torch
import numpy as np
from torch.utils.data import DataLoader
from transformers import AutoTokenizer, AutoModel
from sklearn.preprocessing import LabelEncoder

# Set device (Use CUDA if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load Data
train_df = pd.read_csv("train_with_textual_reports_1.csv")
val_df = pd.read_csv("val_with_textual_reports_1.csv")
test_df = pd.read_csv("test_with_textual_reports_1.csv")

# Extract text and labels
train_texts = train_df["textual_report"].tolist()
val_texts = val_df["textual_report"].tolist()
test_texts = test_df["textual_report"].tolist()

train_labels = train_df["Group"].tolist()
val_labels = val_df["Group"].tolist()
test_labels = test_df["Group"].tolist()

# Encode Labels
label_encoder = LabelEncoder()
label_encoder.fit(train_labels + val_labels + test_labels)

train_label_ids = label_encoder.transform(train_labels)
val_label_ids = label_encoder.transform(val_labels)
test_label_ids = label_encoder.transform(test_labels)

# Load Bio_ClinicalBERT Model
model_name = "emilyalsentzer/Bio_ClinicalBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bioclinicalbert_model = AutoModel.from_pretrained(model_name).to(device)
bioclinicalbert_model.eval()  # Set model to evaluation mode

# Function to Generate Bio_ClinicalBERT Embeddings
def get_bioclinicalbert_embeddings(text_list, tokenizer, model, batch_size=8, max_length=512):
    """
    Encodes a list of texts with Bio_ClinicalBERT,
    returning an [N, hidden_dim] array via mean pooling of the last hidden states.
    """
    all_embeddings = []
    
    # Use DataLoader for efficient batching
    text_loader = DataLoader(text_list, batch_size=batch_size, shuffle=False)
    
    with torch.no_grad():
        for batch_texts in text_loader:
            encodings = tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                max_length=max_length,
                return_tensors="pt"
            ).to(device)  # Move to GPU

            # Get model outputs
            outputs = model(**encodings)
            last_hidden_state = outputs.last_hidden_state  # Shape: [batch_size, seq_len, hidden_dim]

            # Compute mean pooling across seq_len
            batch_emb = last_hidden_state.mean(dim=1)  # Shape: [batch_size, hidden_dim]

            # Convert to numpy and append
            all_embeddings.append(batch_emb.cpu().numpy())

    return np.concatenate(all_embeddings, axis=0)

# Generate Bio_ClinicalBERT Embeddings for Text Reports
train_embeddings = get_bioclinicalbert_embeddings(train_texts, tokenizer, bioclinicalbert_model, batch_size=8, max_length=512)
val_embeddings = get_bioclinicalbert_embeddings(val_texts, tokenizer, bioclinicalbert_model, batch_size=8, max_length=512)
test_embeddings = get_bioclinicalbert_embeddings(test_texts, tokenizer, bioclinicalbert_model, batch_size=8, max_length=512)

# Save embeddings and labels as `.npy` files
np.save("train_text_embeddings_bio_bert.npy", train_embeddings)
np.save("val_text_embeddings_bio_bert.npy", val_embeddings)
np.save("test_text_embeddings_bio_bert.npy", test_embeddings)

np.save("train_text_labels_bio_bert.npy", train_label_ids)
np.save("val_text_labels_bio_bert.npy", val_label_ids)
np.save("test_text_labels_bio_bert.npy", test_label_ids)

print("✅ Bio_ClinicalBERT embeddings extracted and saved successfully!")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import accuracy_score, classification_report

# Convert to PyTorch tensors
train_X = torch.tensor(train_embeddings, dtype=torch.float32)
train_y = torch.tensor(train_labels, dtype=torch.long)
val_X = torch.tensor(val_embeddings, dtype=torch.float32)
val_y = torch.tensor(val_labels, dtype=torch.long)
test_X = torch.tensor(test_embeddings, dtype=torch.float32)
test_y = torch.tensor(test_labels, dtype=torch.long)

# Create TensorDatasets
train_dataset = TensorDataset(train_X, train_y)
val_dataset = TensorDataset(val_X, val_y)
test_dataset = TensorDataset(test_X, test_y)

# Create DataLoaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Determine input dimension from embeddings
input_dim = train_embeddings.shape[1]
num_classes = len(np.unique(train_labels))

# Logistic Regression Model
class LogisticRegressionModel(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(LogisticRegressionModel, self).__init__()
        self.linear = nn.Linear(input_dim, num_classes)
    
    def forward(self, x):
        return self.linear(x)

# Multilayer Perceptron Model
class MLPModel(nn.Module):
    def __init__(self, input_dim, num_classes, hidden_dim=128):
        super(MLPModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, num_classes)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

def train_model(model, train_loader, val_loader, num_epochs=10, learning_rate=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    for epoch in range(num_epochs):
        model.train()
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        val_preds = []
        val_targets = []
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                _, preds = torch.max(outputs, 1)
                val_preds.extend(preds.cpu().numpy())
                val_targets.extend(batch_y.cpu().numpy())

        val_accuracy = accuracy_score(val_targets, val_preds)
        print(f"Epoch {epoch+1}/{num_epochs}, Validation Accuracy: {val_accuracy:.4f}")

# Logistic Regression
print("Training Logistic Regression Model...")
lr_model = LogisticRegressionModel(input_dim, num_classes)
train_model(lr_model, train_loader, val_loader)

# Multilayer Perceptron
print("\nTraining MLP Model...")
mlp_model = MLPModel(input_dim, num_classes)
train_model(mlp_model, train_loader, val_loader)

def evaluate_model(model, test_loader):
    model.eval()
    test_preds = []
    test_targets = []
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    with torch.no_grad():
        for batch_X, batch_y in test_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            outputs = model(batch_X)
            _, preds = torch.max(outputs, 1)
            test_preds.extend(preds.cpu().numpy())
            test_targets.extend(batch_y.cpu().numpy())

    test_accuracy = accuracy_score(test_targets, test_preds)
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print("Classification Report:")
    print(classification_report(test_targets, test_preds))

# Evaluate Logistic Regression Model
print("Evaluating Logistic Regression Model...")
evaluate_model(lr_model, test_loader)

# Evaluate MLP Model
print("\nEvaluating MLP Model...")
evaluate_model(mlp_model, test_loader)
```

```python
import os
import pandas as pd
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AdamW, get_linear_schedule_with_warmup
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report  # For generating classification report

# Set device (Use CUDA if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load Data
train_df = pd.read_csv("train_with_textual_reports_1.csv")
val_df = pd.read_csv("val_with_textual_reports_1.csv")
test_df = pd.read_csv("test_with_textual_reports_1.csv")

# Extract text and labels
train_texts = train_df["textual_report"].tolist()
val_texts = val_df["textual_report"].tolist()
test_texts = test_df["textual_report"].tolist()

train_labels = train_df["Group"].tolist()
val_labels = val_df["Group"].tolist()
test_labels = test_df["Group"].tolist()

# Encode Labels
label_encoder = LabelEncoder()
label_encoder.fit(train_labels + val_labels + test_labels)
train_label_ids = label_encoder.transform(train_labels)
val_label_ids = label_encoder.transform(val_labels)
test_label_ids = label_encoder.transform(test_labels)
num_classes = len(label_encoder.classes_)

# Load Bio_ClinicalBERT for sequence classification
model_name = "emilyalsentzer/Bio_ClinicalBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_classes)
model.to(device)

# Create a custom dataset class
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        # Tokenize text
        encoding = self.tokenizer(
            text,
            padding='max_length',
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        # Remove the batch dimension (squeeze) and add the label
        item = {key: val.squeeze(0) for key, val in encoding.items()}
        item['labels'] = torch.tensor(label, dtype=torch.long)
        return item

# Create DataLoaders
batch_size = 16
train_dataset = TextDataset(train_texts, train_label_ids, tokenizer, max_length=512)
val_dataset = TextDataset(val_texts, val_label_ids, tokenizer, max_length=512)
test_dataset = TextDataset(test_texts, test_label_ids, tokenizer, max_length=512)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Define optimizer and scheduler
epochs = 5  # Set the number of epochs
optimizer = AdamW(model.parameters(), lr=2e-5)
total_steps = len(train_loader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Training and validation loop
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}/{epochs}")
    
    # Training phase
    model.train()
    train_loss_total = 0
    train_preds = []
    train_labels_all = []
    
    for batch in train_loader:
        # Move batch to device
        inputs = {key: val.to(device) for key, val in batch.items()}
        
        # Forward pass; model returns loss when labels are provided
        outputs = model(**inputs)
        loss = outputs.loss
        train_loss_total += loss.item()
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
        
        # Collect predictions and labels for training accuracy
        logits = outputs.logits
        preds = torch.argmax(logits, dim=-1)
        train_preds.extend(preds.cpu().numpy())
        train_labels_all.extend(inputs['labels'].cpu().numpy())
    
    # Calculate training metrics
    train_loss_avg = train_loss_total / len(train_loader)
    train_accuracy = np.mean(np.array(train_preds) == np.array(train_labels_all))
    
    # Validation phase
    model.eval()
    val_loss_total = 0
    val_preds = []
    val_labels_all = []
    
    with torch.no_grad():
        for batch in val_loader:
            inputs = {key: val.to(device) for key, val in batch.items()}
            outputs = model(**inputs)
            loss = outputs.loss
            val_loss_total += loss.item()
            
            logits = outputs.logits
            preds = torch.argmax(logits, dim=-1)
            val_preds.extend(preds.cpu().numpy())
            val_labels_all.extend(inputs['labels'].cpu().numpy())
    
    val_loss_avg = val_loss_total / len(val_loader)
    val_accuracy = np.mean(np.array(val_preds) == np.array(val_labels_all))
    
    # Generate classification report on validation set
    val_class_report = classification_report(val_labels_all, val_preds, target_names=label_encoder.classes_, digits=4)
    
    # Print metrics for this epoch
    print(f"Training Loss: {train_loss_avg:.4f} | Training Accuracy: {train_accuracy:.4f}")
    print(f"Validation Loss: {val_loss_avg:.4f} | Validation Accuracy: {val_accuracy:.4f}")
    print("\nValidation Classification Report:")
    print(val_class_report)

# Test evaluation phase (run after training)
model.eval()
test_loss_total = 0
test_preds = []
test_labels_all = []

with torch.no_grad():
    for batch in test_loader:
        inputs = {key: val.to(device) for key, val in batch.items()}
        outputs = model(**inputs)
        loss = outputs.loss
        test_loss_total += loss.item()
        
        logits = outputs.logits
        preds = torch.argmax(logits, dim=-1)
        test_preds.extend(preds.cpu().numpy())
        test_labels_all.extend(inputs['labels'].cpu().numpy())

test_loss_avg = test_loss_total / len(test_loader)
test_accuracy = np.mean(np.array(test_preds) == np.array(test_labels_all))
test_class_report = classification_report(test_labels_all, test_preds, target_names=label_encoder.classes_, digits=4)

print(f"\nTest Loss: {test_loss_avg:.4f} | Test Accuracy: {test_accuracy:.4f}")
print("\nTest Classification Report:")
print(test_class_report)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

# -------------------------
# Dataset with Missing Mask
# -------------------------
class TabularDatasetWithMask(Dataset):
    def __init__(self, X, mask, y):
        """
        X: numpy array of shape (num_samples, num_features)
        mask: numpy array of same shape, 1 for observed and 0 for missing
        y: numpy array of target labels (for classification)
        """
        self.X = torch.tensor(X, dtype=torch.float32)
        self.mask = torch.tensor(mask, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.mask[idx], self.y[idx]

# -------------------------
# Simplified NAIM Model
# -------------------------
class NAIM(nn.Module):
    def __init__(self, num_features, hidden_dim, num_classes, num_layers=2, dropout=0.1):
        """
        num_features: number of features (each treated as a token)
        hidden_dim: dimension of each token’s embedding space
        num_classes: number of output classes
        num_layers: number of transformer encoder layers
        dropout: dropout probability in transformer
        """
        super(NAIM, self).__init__()
        self.num_features = num_features
        # Create an embedding (a simple linear layer) for each feature.
        self.feature_embeddings = nn.ModuleList([nn.Linear(1, hidden_dim) for _ in range(num_features)])
        # Learnable positional encoding (one per feature)
        self.positional_encoding = nn.Parameter(torch.randn(num_features, hidden_dim))
        # Transformer encoder (expects input shape: [sequence_length, batch_size, hidden_dim])
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        # Final classifier: aggregate token representations (here by mean pooling) then map to logits.
        self.output_layer = nn.Linear(hidden_dim, num_classes)
    
    def forward(self, x, mask=None):
        """
        x: tensor of shape (batch_size, num_features)
        mask: tensor of shape (batch_size, num_features); 1 if observed, 0 if missing.
        
        For missing features we simply replace the value by 0. In a more advanced version, you might
        add a dedicated missing embedding.
        """
        batch_size = x.size(0)
        embedded_tokens = []
        # Process each feature (token) separately.
        for i in range(self.num_features):
            # Get the i-th feature for all samples: shape (batch_size, 1)
            xi = x[:, i].unsqueeze(1)
            if mask is not None:
                # For missing entries (mask==0), set value to 0 (a placeholder)
                missing_i = (mask[:, i] == 0).unsqueeze(1).float()
                xi = xi * (1 - missing_i)
            # Apply the dedicated embedding layer for feature i.
            emb_i = self.feature_embeddings[i](xi)  # (batch_size, hidden_dim)
            # Add as a token (unsqueeze to have shape (batch_size, 1, hidden_dim))
            embedded_tokens.append(emb_i.unsqueeze(1))
        
        # Concatenate tokens along the feature dimension: (batch_size, num_features, hidden_dim)
        x_emb = torch.cat(embedded_tokens, dim=1)
        # Add positional encoding (broadcasted over the batch)
        x_emb = x_emb + self.positional_encoding.unsqueeze(0)
        # Transformer expects shape (sequence_length, batch_size, hidden_dim)
        x_trans = x_emb.transpose(0, 1)
        # Pass through the transformer encoder.
        x_encoded = self.transformer_encoder(x_trans)
        # Aggregate the token representations by taking the mean over the tokens.
        x_agg = x_encoded.mean(dim=0)
        # Compute logits.
        logits = self.output_layer(x_agg)
        return logits

# -------------------------
# Main Training & Evaluation
# -------------------------
if __name__ == '__main__':
    # For reproducibility
    np.random.seed(42)
    torch.manual_seed(42)
    
    # Generate synthetic tabular data
    num_samples = 1000
    num_features = 10
    X = np.random.randn(num_samples, num_features)
    # Create a binary target: 1 if sum > 0, else 0.
    y = (np.sum(X, axis=1) > 0).astype(int)
    
    # Introduce missing values randomly
    missing_rate = 0.3
    mask = (np.random.rand(num_samples, num_features) > missing_rate).astype(int)
    # Replace missing entries with 0 (placeholder; model knows these were missing via the mask)
    X[mask == 0] = 0.0
    
    # Split into training and testing sets (80/20 split)
    split = int(0.8 * num_samples)
    X_train, X_test = X[:split], X[split:]
    mask_train, mask_test = mask[:split], mask[split:]
    y_train, y_test = y[:split], y[split:]
    
    # Create dataset and dataloaders
    train_dataset = TabularDatasetWithMask(X_train, mask_train, y_train)
    test_dataset = TabularDatasetWithMask(X_test, mask_test, y_test)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Instantiate the NAIM model
    hidden_dim = 32
    num_classes = 2
    model = NAIM(num_features=num_features, hidden_dim=hidden_dim, num_classes=num_classes)
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    # Define loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    
    # Training loop
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for batch in train_loader:
            X_batch, mask_batch, y_batch = batch
            X_batch, mask_batch, y_batch = X_batch.to(device), mask_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch, mask=mask_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * X_batch.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
    
    # Evaluation on the test set
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in test_loader:
            X_batch, mask_batch, y_batch = batch
            X_batch, mask_batch, y_batch = X_batch.to(device), mask_batch.to(device), y_batch.to(device)
            outputs = model(X_batch, mask=mask_batch)
            _, predicted = torch.max(outputs, dim=1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
    accuracy = correct / total
    print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

# -------------------------
# Dataset: keep data natural, add missing indicator and replace NaNs with -999
# -------------------------
class TabularDatasetWithMissingIndicator(Dataset):
    def __init__(self, X, y):
        """
        X: numpy array of shape (num_samples, num_features) with natural NaNs where data is missing.
        y: numpy array of target labels (for classification)
        """
        self.X = X
        self.y = y
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        # Get the original row (which may have NaNs)
        x_row = self.X[idx]
        # Compute missing indicator: 1 if missing, 0 if observed.
        missing_indicator = np.isnan(x_row).astype(np.float32)
        # Replace NaNs with the sentinel value (-999)
        x_filled = np.where(np.isnan(x_row), -999, x_row).astype(np.float32)
        return (
            torch.tensor(x_filled, dtype=torch.float32),
            torch.tensor(missing_indicator, dtype=torch.float32),
            torch.tensor(self.y[idx], dtype=torch.long)
        )

# -------------------------
# Simplified NAIM Model with Missing Embeddings
# -------------------------
class NAIM(nn.Module):
    def __init__(self, num_features, hidden_dim, num_classes, num_layers=2, dropout=0.1):
        """
        num_features: number of features (each treated as a token)
        hidden_dim: dimension of each token’s embedding space
        num_classes: number of output classes
        num_layers: number of transformer encoder layers
        dropout: dropout probability in transformer
        """
        super(NAIM, self).__init__()
        self.num_features = num_features
        # Create an embedding (a simple linear layer) for each feature.
        self.feature_embeddings = nn.ModuleList([
            nn.Linear(1, hidden_dim) for _ in range(num_features)
        ])
        # Learnable embedding for missing values for each feature.
        self.missing_embeddings = nn.ParameterList([
            nn.Parameter(torch.randn(hidden_dim)) for _ in range(num_features)
        ])
        # Learnable positional encoding (one per feature)
        self.positional_encoding = nn.Parameter(torch.randn(num_features, hidden_dim))
        # Transformer encoder (expects input shape: [sequence_length, batch_size, hidden_dim])
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        # Final classifier: aggregate token representations (mean pooling) then map to logits.
        self.output_layer = nn.Linear(hidden_dim, num_classes)
    
    def forward(self, x, missing_indicator):
        """
        x: tensor of shape (batch_size, num_features) containing feature values,
           where missing values have been replaced by a sentinel (-999).
        missing_indicator: tensor of shape (batch_size, num_features), with 1 if the value was missing.
        """
        batch_size = x.size(0)
        embedded_tokens = []
        # Process each feature as a separate token.
        for i in range(self.num_features):
            # Extract feature i for all samples; shape: (batch_size, 1)
            xi = x[:, i].unsqueeze(1)
            # Get the corresponding embedding for the value.
            emb_i = self.feature_embeddings[i](xi)  # shape: (batch_size, hidden_dim)
            # Get the missing indicator for feature i (1 if missing, 0 if observed)
            missing_i = missing_indicator[:, i].unsqueeze(1)  # shape: (batch_size, 1)
            # Add the missing embedding if the value is missing.
            missing_emb = self.missing_embeddings[i].unsqueeze(0)  # shape: (1, hidden_dim)
            emb_i = emb_i + missing_i * missing_emb
            # Append token (unsqueeze to have shape (batch_size, 1, hidden_dim))
            embedded_tokens.append(emb_i.unsqueeze(1))
        
        # Concatenate tokens along the feature dimension: shape (batch_size, num_features, hidden_dim)
        x_emb = torch.cat(embedded_tokens, dim=1)
        # Add positional encoding (broadcast over batch dimension)
        x_emb = x_emb + self.positional_encoding.unsqueeze(0)
        # Transformer expects input of shape (sequence_length, batch_size, hidden_dim)
        x_trans = x_emb.transpose(0, 1)
        # Process through transformer encoder.
        x_encoded = self.transformer_encoder(x_trans)
        # Aggregate token representations (mean pooling over the token dimension)
        x_agg = x_encoded.mean(dim=0)
        # Produce final classification logits.
        logits = self.output_layer(x_agg)
        return logits

# -------------------------
# Main Training & Evaluation
# -------------------------
if __name__ == '__main__':
    # For reproducibility
    np.random.seed(42)
    torch.manual_seed(42)
    
    # Generate synthetic tabular data
    num_samples = 1000
    num_features = 10
    X = np.random.randn(num_samples, num_features)
    # Create a binary target: 1 if sum > 0, else 0.
    y = (np.sum(X, axis=1) > 0).astype(int)
    
    # Introduce missing values randomly (simulate missingness)
    missing_rate = 0.3
    # Create a boolean mask where True means observed; False means missing.
    mask = (np.random.rand(num_samples, num_features) > missing_rate)
    # Instead of replacing missing entries with a “default” (e.g., 0), set them to NaN:
    X[~mask] = np.nan  # leave the data in its original (natural) form.
    
    # Split into training and testing sets (80/20 split)
    split = int(0.8 * num_samples)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    
    # Create datasets and dataloaders. The dataset __getitem__ will compute the missing indicator
    # and replace NaNs with the sentinel value (-999) on the fly.
    train_dataset = TabularDatasetWithMissingIndicator(X_train, y_train)
    test_dataset = TabularDatasetWithMissingIndicator(X_test, y_test)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Instantiate the NAIM model.
    hidden_dim = 32
    num_classes = 2
    model = NAIM(num_features=num_features, hidden_dim=hidden_dim, num_classes=num_classes)
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    # Define loss and optimizer.
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    
    # Training loop.
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for batch in train_loader:
            X_batch, missing_batch, y_batch = batch
            X_batch = X_batch.to(device)
            missing_batch = missing_batch.to(device)
            y_batch = y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch, missing_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * X_batch.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
    
    # Evaluation on the test set.
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in test_loader:
            X_batch, missing_batch, y_batch = batch
            X_batch = X_batch.to(device)
            missing_batch = missing_batch.to(device)
            y_batch = y_batch.to(device)
            outputs = model(X_batch, missing_batch)
            _, predicted = torch.max(outputs, dim=1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
    accuracy = correct / total
    print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

# -------------------------
# Dataset: leave data in original form (with NaNs)
# -------------------------
class TabularDataset(Dataset):
    def __init__(self, X, y):
        """
        X: numpy array of shape (num_samples, num_features) with np.nan representing missing values.
        y: numpy array of target labels.
        """
        self.X = X.astype(np.float32)  # keep NaNs as np.nan
        self.y = y.astype(np.int64)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        # Return the row as-is, with NaNs intact.
        x_row = self.X[idx]
        return torch.tensor(x_row), torch.tensor(self.y[idx])

# -------------------------
# NAIM Model that handles NaNs internally
# -------------------------
class NAIM(nn.Module):
    def __init__(self, num_features, hidden_dim, num_classes, num_layers=2, dropout=0.1):
        """
        num_features: Number of features (each treated as a token)
        hidden_dim: Dimension of each token’s embedding space
        num_classes: Number of output classes
        num_layers: Number of transformer encoder layers
        dropout: Dropout probability in the transformer
        """
        super(NAIM, self).__init__()
        self.num_features = num_features
        # Create a dedicated embedding (a linear layer) for each feature.
        self.feature_embeddings = nn.ModuleList([
            nn.Linear(1, hidden_dim) for _ in range(num_features)
        ])
        # Create a learnable missing embedding for each feature.
        self.missing_embeddings = nn.ParameterList([
            nn.Parameter(torch.randn(hidden_dim)) for _ in range(num_features)
        ])
        # Learnable positional encoding for each feature.
        self.positional_encoding = nn.Parameter(torch.randn(num_features, hidden_dim))
        # Transformer encoder (expects input shape: [sequence_length, batch_size, hidden_dim])
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        # Final classifier: aggregate token representations (mean pooling) then map to logits.
        self.output_layer = nn.Linear(hidden_dim, num_classes)
    
    def forward(self, x):
        """
        x: Tensor of shape (batch_size, num_features) which may contain NaNs.
        For each feature, if the value is NaN the model will use the dedicated missing embedding.
        """
        batch_size = x.size(0)
        embedded_tokens = []
        for i in range(self.num_features):
            # Extract the i-th feature for all samples (shape: [batch_size, 1]).
            xi = x[:, i].unsqueeze(1)
            # Create a mask: True where xi is NaN.
            mask = torch.isnan(xi)
            # For passing xi to the linear layer, replace NaNs with 0.
            xi_filled = torch.where(mask, torch.zeros_like(xi), xi)
            emb_val = self.feature_embeddings[i](xi_filled)  # shape: [batch_size, hidden_dim]
            # Get the learnable missing embedding for feature i.
            missing_emb = self.missing_embeddings[i].unsqueeze(0).expand(batch_size, -1)
            # Expand the mask to match the embedding dimension.
            mask_expanded = mask.expand(-1, emb_val.size(1))
            # For each sample, if the value is missing (True in mask) use missing_emb; otherwise use emb_val.
            emb_i = torch.where(mask_expanded, missing_emb, emb_val)
            # Add a new token dimension.
            embedded_tokens.append(emb_i.unsqueeze(1))
        
        # Concatenate tokens along the feature dimension -> shape: [batch_size, num_features, hidden_dim].
        x_emb = torch.cat(embedded_tokens, dim=1)
        # Add positional encoding.
        x_emb = x_emb + self.positional_encoding.unsqueeze(0)
        # Transpose for transformer (sequence_length, batch_size, hidden_dim).
        x_trans = x_emb.transpose(0, 1)
        # Process through the transformer encoder.
        x_encoded = self.transformer_encoder(x_trans)
        # Aggregate token representations (mean pooling over tokens).
        x_agg = x_encoded.mean(dim=0)
        # Produce classification logits.
        logits = self.output_layer(x_agg)
        return logits

# -------------------------
# Main Training & Evaluation
# -------------------------
if __name__ == '__main__':
    np.random.seed(42)
    torch.manual_seed(42)
    
    # Generate synthetic tabular data.
    num_samples = 1000
    num_features = 10
    X = np.random.randn(num_samples, num_features)
    # Create a binary target: 1 if sum > 0, else 0.
    y = (np.sum(X, axis=1) > 0).astype(np.int64)
    
    # Introduce missing values randomly: leave missing entries as np.nan.
    missing_rate = 0.3
    mask = np.random.rand(num_samples, num_features) > missing_rate
    X[~mask] = np.nan  # keep data natural
    
    # Split into training and testing sets (80/20 split).
    split = int(0.8 * num_samples)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    
    # Create datasets and dataloaders.
    train_dataset = TabularDataset(X_train, y_train)
    test_dataset = TabularDataset(X_test, y_test)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Instantiate the NAIM model.
    hidden_dim = 32
    num_classes = 2
    model = NAIM(num_features=num_features, hidden_dim=hidden_dim, num_classes=num_classes)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    # Define loss and optimizer.
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    
    # Training loop.
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * X_batch.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
    
    # Evaluation on the test set.
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)
            _, predicted = torch.max(outputs, dim=1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
    accuracy = correct / total
    print(f"Test Accuracy: {accuracy:.4f}")
```
